```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
studentData= pd.read_csv("batch12.csv")
studentData

```

```python

 non_zero = [	'Subject1',	'Subject2',	'Subject3',	'Subject4',	'Subject5',	'Subject6',	'Day',	'Subject']
 #,'Predict']
 for col in non_zero:
   studentData[col] = studentData[col].replace(0, np.NaN)
   mean = int(studentData[col].median(skipna=True))
   studentData[col] = studentData[col].replace(np.NaN, mean)
```

```python
# Assuming studentData is your main dataset and it has a target column named 'Target'
# Split the data into training and test sets
train = studentData.sample(frac=0.7, random_state=1000)
test = studentData.drop(train.index)

# Extract the subject columns for training and test sets
subject_columns = ['Subject1', 'Subject2', 'Subject3', 'Subject4', 'Subject5', 'Subject6','Day','Subject']
train_subjects = train[subject_columns]
test_subjects = test[subject_columns]

# Assuming 'Target' is the column you want to predict
#problem heeereee
train_target = train['Day']

# Fit the KNN model
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(train_subjects, train_target)

# Predict the target for the test set
predicted_schedule = knn.predict(test_subjects)
predicted_schedule_df = pd.DataFrame(predicted_schedule, columns=['Predict'])

# Print the predicted schedule
print(predicted_schedule_df)
```

```python
mean_squared_error(test['Day'], predicted_schedule)
```
