```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
studentData= pd.read_csv("batch12.csv")
studentData2=pd.read_csv("batch2.csv")
studentData3=pd.read_csv("batch3.csv")
studentData4=pd.read_csv("batch4.csv")
studentData5=pd.read_csv("batch5.csv")
studentData4

```

```python
non_zero = ['Subject1', 'Subject2', 'Subject3', 'Subject4', 'Subject5', 'Subject6', 'Day', 'Subject']
for col in non_zero:
    studentData[col] = studentData[col].replace(0, np.NaN)
    mean = studentData[col].fillna(0).median(skipna=True)  # Replace NaN with 0 before calculating the median
    studentData[col] = studentData[col].replace(np.NaN, mean)

non_zero = ['Subject1', 'Subject2', 'Subject3', 'Subject4', 'Subject5', 'Subject6', 'Day', 'Subject']
for col in non_zero:
    studentData2[col] = studentData2[col].replace(0, np.NaN)
    mean = studentData2[col].fillna(0).median(skipna=True)  # Replace NaN with 0 before calculating the median
    studentData2[col] = studentData2[col].replace(np.NaN, mean)

non_zero = ['Subject1', 'Subject2', 'Subject3', 'Subject4', 'Subject5', 'Subject6', 'Day', 'Subject']
for col in non_zero:
    studentData3[col] = studentData3[col].replace(0, np.NaN)
    mean = studentData3[col].fillna(0).median(skipna=True)  # Replace NaN with 0 before calculating the median
    studentData3[col] = studentData3[col].replace(np.NaN, mean)

non_zero = ['Subject1', 'Subject2', 'Subject3', 'Subject4', 'Subject5', 'Subject6', 'Predict day', 'Predict subject']
for col in non_zero:
    studentData4[col] = studentData4[col].replace(0, np.NaN)
    mean = studentData4[col].fillna(0).median(skipna=True)  # Replace NaN with 0 before calculating the median
    studentData4[col] = studentData4[col].replace(np.NaN, mean)

non_zero = ['Subject1', 'Subject2', 'Subject3', 'Subject4', 'Subject5', 'Subject6', 'Predict day', 'Predict subject']
for col in non_zero:
    studentData5[col] = studentData5[col].replace(0, np.NaN)
    mean = studentData5[col].fillna(0).median(skipna=True)  # Replace NaN with 0 before calculating the median
    studentData5[col] = studentData5[col].replace(np.NaN, mean)
   
```
